**PIE CHART**

```python
x = [10,20,30,40]
y= ['c','c++','java','Python']

plt.figure(figsize=(16,8))
plt.pie(x,labels=y)
plt.show()
```

```python
x = [10,20,30,40]
y= ['c','c++','java','Python']
ex = [0.0,0.4,0.0,0.0]

plt.figure(figsize=(16,8))
plt.pie(x,labels=y,explode=ex)
plt.show()

# -> explode function 
```

```python
x = [10,20,30,40]
y= ['c','c++','java','Python']
ex = [0.0,0.4,0.0,0.0]

plt.figure(figsize=(20,10))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%')
plt.show()

# -> showing the percentage using autopct 
```

```python
x = [10,20,30,40]
y= ['c','c++','java','Python']
ex = [0.0,0.4,0.0,0.0]

plt.figure(figsize=(20,10))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',radius=1.5)
plt.savefig('PIE Chart PLOT')
plt.show()

# -> changed the radius 
```

```python
# -> changing of starting angle 
x = [10,20,30,40]
y= ['c','c++','java','Python']
ex = [0.0,0.4,0.0,0.0]

plt.figure(figsize=(20,10))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',radius=1.2,startangle=45)
plt.show()
```

```python
# -> font size of the text 

x = [10,20,30,40]
y= ['c','c++','java','Python']
ex = [0.0,0.4,0.0,0.0]

plt.figure(figsize=(20,10))
plt.pie(x,labels=y,explode=ex,autopct='%0.1f%%',radius=1.2,textprops={'fontsize': 30})
plt.show()
```

**STEM PLOTS**
* used for showing signals

```python
x = [1,2,3,4,5,6]
y=[10,20,30,40,50,60]


plt.figure(figsize=(18,9))
plt.stem(x,y,linefmt='--',markerfmt='rv')
plt.show()
```

```python
# -> Bottom parameter 

x = [1,2,3,4,5,6]
y=[10,20,30,40,50,60]


plt.figure(figsize=(18,9))
plt.stem(x,y,linefmt='--',markerfmt='rv',bottom=1)
plt.show()
```

```python
# changing base line 

x = [1,2,3,4,5,6]
y=[10,20,30,40,50,60]


plt.figure(figsize=(18,9))
plt.stem(x,y,linefmt='--',markerfmt='rv',bottom=1,basefmt='g')
plt.show()
```

**STACK PLOT / AREA PLOT**

```python
x = [1,2,3,4,5,6]
y=[1,2,3,4,5,6]


plt.figure(figsize=(18,9))
plt.stackplot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y=[2,2,9,4,15,10]


plt.figure(figsize=(18,9))
plt.stackplot(x,y)
plt.show()
```

```python
x = [1,2,3,4,5,6]
y=[2,10,9,4,15,10]
y1=[2,4,11,19,17,5]
y2=[2,8,11,19,17,5]


plt.figure(figsize=(18,9))
plt.stackplot(x,y,y1,y2)
plt.show()
```

```python
# -> labels 

x = [1,2,3,4,5,6]
y=[2,10,9,4,15,10]
y1=[2,4,11,19,17,5]
y2=[2,8,11,19,17,5]

l = ['piyal','shakir','mahir']


plt.figure(figsize=(18,9))
plt.stackplot(x,y,y1,y2,labels=l)
plt.legend(loc=2)
plt.show()
```

**STEP PLOT**

```python
x = [1,2,3,4,5,6]
y=[2,10,9,4,15,10]

plt.figure(figsize=(19,8))
plt.step(x,y)
plt.savefig("Step Plot Graph")
plt.show()
```
